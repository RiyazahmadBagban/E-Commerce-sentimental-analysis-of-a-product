```python
df_clean_text=df1.Reviews.str.split()
df_clean_text
```

```python
import nltk
nltk.download('stopwords')
sw=nltk.corpus.stopwords.words('english')
sw.remove('not')
```

```python
#  nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer 
ls=WordNetLemmatizer()
lemma = df_clean_text.apply(lambda x: [ls.lemmatize(word) for word in x if word not in sw])
lemma = lemma.apply(lambda x: ' '.join(x))
df1['Lemmas']=lemma.to_frame()
```

```python
from textblob import TextBlob
df1['polarity']=(round(df1['Lemmas'].apply(lambda x:TextBlob(x).sentiment.polarity),4))
```

```python
 df1
```

```python
df1=df1[['Lemmas','polarity']]
```

```python
df1
```

```python
polar=[]
for i in df1.polarity:
    if i==0.0:
        polar.append('neutral')
    elif i>0.1:
        polar.append('positive')
    else:
        polar.append('negative')
```

```python
df1['sentiment']=polar
```

```python
df1.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
encoded = le.fit_transform(df1["sentiment"])
df1["vad_encoded_value"] = encoded
```

```python
#TFIDF - Term frequency inverse Document Frequencyt
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english', max_features= 1000, max_df = 0.5, smooth_idf=True) #keep top 1000 words
doc_vec = vectorizer.fit_transform(df1["Lemmas"])
x=doc_vec
y=df1['vad_encoded_value']
```

```python
from imblearn.over_sampling import SMOTE
```

```python
X_resampled, y_resampled = SMOTE().fit_resample(x, y)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_resampled,y_resampled,train_size=0.75,random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB as MB
# Multinomial Naive Bayes
classifier_mb = MB()
cm=classifier_mb.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
y_pred_MB=cm.predict(X_test)
accuracy_score(y_test,y_pred_MB)
```

```python
MB_acc=round(accuracy_score(y_test,y_pred_MB),4)
```

```python
from sklearn.linear_model import LogisticRegression
classifier_model_LR=LogisticRegression(random_state=1)
LR=classifier_model_LR.fit(X_train,y_train)

y_pred_LR=LR.predict(X_test)
accuracy_score(y_test,y_pred_LR)
```

```python
LR_acc=round(accuracy_score(y_test,y_pred_LR),4)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier_model_DT=DecisionTreeClassifier(random_state=1,max_depth=15)
dt=classifier_model_DT.fit(X_train,y_train)

y_pred_DT=dt.predict(X_test)
accuracy_score(y_test,y_pred_DT)
```

```python
DT_acc=round(accuracy_score(y_test,y_pred_DT),4)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier_model_RF=RandomForestClassifier(random_state=1)
rf=classifier_model_RF.fit(X_train,y_train)

y_pred_RF=rf.predict(X_test)
accuracy_score(y_test,y_pred_RF)
```

```python
RF_acc=round(accuracy_score(y_test,y_pred_RF),4)
```

```python
from sklearn.ensemble import AdaBoostClassifier
classifier_model_ADA=AdaBoostClassifier(random_state=1)
ada=classifier_model_ADA.fit(X_train,y_train)
y_pred_ADA=ada.predict(X_test)
accuracy_score(y_test,y_pred_ADA)
```

```python
ADA_acc=round(accuracy_score(y_test,y_pred_ADA),4)
```

```python
data=[['Multinomial Naive Bayes',MB_acc],['Logistic Regression',LR_acc],['Decision Tree',DT_acc],['Random Forest',RF_acc],
      ['Ada Boost',ADA_acc]]

df=pd.DataFrame(data,columns=['Model','test_accuracy'])
```

```python
import re
import string
from nltk.corpus import stopwords
# from nltk.stem import WordNetLemmatizer 
# ls = WordNetLemmatizer()

def no_punc(text):
    docs = []
    text=text.lower()
    text=re.sub('[^A-Za-z]+', ' ',text)
    text = text.split()
    text=[ls.lemmatize(word) for word in text if word not in sw]
    text=' '.join(text)
    docs.append(text)
    return docs
```

```python
text_clean=no_punc(testing) # list type
```

```python
df2=vectorizer.transform(text_clean).toarray()
pred=cm.predict(df2)
label=pred[0]

if label==2:
    print('Positive')
elif label==0:
    print('Negative')
else:
    print('Neutral')
```

```python
from pickle import dump
dump(vectorizer,open('vectorizer.pkl','wb'))
dump(cm,open('model.pkl','wb'))
```

```python
from pickle import load

def predict(text):
    vector=load(open('vectorizer.pkl','rb'))
    classify=load(open('model.pkl','rb'))
    clean_text=no_punc(text)
    vector_clean_text=vector.transform(clean_text)
    vector_array=vector_clean_text.toarray()
    prediction=classify.predict(vector_array)
    return prediction
```
