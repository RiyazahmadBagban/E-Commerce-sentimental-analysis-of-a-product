```python
from bs4 import BeautifulSoup as bs
import requests
link='https://www.amazon.in/Tribit-Bluetooth-Headphones-Lightweight-Lightening/product-reviews/B07W5KV2LM/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'

page=requests.get(link) # 200 is a standard response for successful HTTP requests and 404 tells that the requested resource could not be found.
# page
page.content
soup=bs(page.content,'html.parser')#Fundamentally, parsing is necessary because different entities need the data to be in different forms. 
                            #Parsing allows transforming data in a way that can be understood by a specific software. The obvious example is programs ‚Äî they are written by humans, but they must be executed by computers
# soup
#print(soup.prettify())#The prettify() method will turn a Beautiful Soup parse tree into a nicely formatted Unicode string,

name=soup.find_all('span',class_="a-profile-name")
cust_name=[]
for i in range(0,len(name)):
    cust_name.append(name[i].get_text())

title=soup.find_all('a',class_='review-title-content')
review_title=[]
for i in range(0,len(title)):
    review_title.append(title[i].get_text())
    
    
review_title[:]=[title.lstrip('\n') for title in review_title]# l n(left and  right) strip means rmoving perticuler argument in string here is \n
review_title[:]=[title.rstrip('\n') for title in review_title]

#cust_name.pop(0) # if ypu want exact index in every column you can use or else leave it , at on one time( you will get  error about index error at data frame creation there you will do DF in series)

rate= soup.find_all('i',class_='review-rating')
rating=[]
for i in range(0,len(rate)):
    rating.append(rate[i].get_text())
    
    
cust_review=soup.find_all('span',{'data-hook':'review-body'})
review=[]
for i in range(0,len(cust_review)):
    review.append(cust_review[i].get_text())
    
review[:]=[cust_review.lstrip('\n') for cust_review in review]
review[:]=[cust_review.rstrip('\n') for cust_review in review]

# len(cust_name)
# len(review_title)
# len(rating)
# len(review)

import pandas as pd
df=pd.DataFrame()
df
df['CustomerName']=pd.Series(cust_name)
df['ReviewTitle']=pd.Series(review_title)
df['Ratings']=pd.Series(rating)
df['Review']=pd.Series(review)
df
# df.to_csv(r'D:\D S\for project\product_reviews.csv',index=True) # true indicate index in excel sheet, if false index column will not appear

 
```
