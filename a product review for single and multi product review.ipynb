```python
from bs4 import BeautifulSoup as bs
import requests
from time import sleep
from random import randint
import numpy as np

#Declaring the headers 
headers = {"Accept-Language": "en-US,en;q=0.5"}

cust_name=[]
review=[]
rating=[]

#creating an array of values and passing it in the url for dynamic webpages
pages =np.arange(1,150)

#the whole core of the script
for page in pages:
    page=requests.get('https://www.amazon.in/Pampers-Large-Diapers-Pants-Count/product-reviews/B07CXFJT8Q/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews&pageNumber='+str(page))                     
    soup = bs(page.content, 'html.parser') 
    
    names = soup.findAll('div',attrs={'class': 'a-section celwidget'})
    sleep(randint(2,8))
    for nm in names:
        name=nm.a.text
        cust_name.append(name)
        
        
    
    rate= soup.find_all('i',class_='review-rating')
    for i in range(len(rate)):
        rating.append(rate[i].get_text()) 
        
    cust_review=soup.find_all('div',attrs={'class': 'a-row a-spacing-small review-data'})
    for i in range(len(cust_review)):
        review.append(cust_review[i].get_text())
        
        review[:]=[cust_review.lstrip('\n') for cust_review in review]
        review[:]=[cust_review.rstrip('\n') for cust_review in review]


     

        
# rating    
# rating
# len(review)
```

```python
import pandas as pd
df=pd.DataFrame()
df  

df['CustomerName']=pd.Series(cust_name)
df['Ratings']=pd.Series(rating)
df['Reviews']=pd.Series(review)
df
```

```python
df.to_csv(r'D:\D S\for project\final_product_reviews.csv',index=True)
```
